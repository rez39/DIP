```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import math 
import time
import cmath
from scipy.signal import convolve2d
NaN=np.NaN
from statistics import median 
```

```python
#question 7
def rotImage(img,angle,retain):
    m,n=img.shape
    A=np.array([[np.cos(np.deg2rad(angle)),-np.sin(np.deg2rad(angle))],[np.sin(np.deg2rad(angle)),np.cos(np.deg2rad(angle))]])
    print(A)
    if retain==True:
        output=np.zeros([m,n])
        for i in range(m):
            for j in range(n):
                x=np.matmul(A,np.array([[i-math.floor(m/2)],[j-math.floor(n/2)]]))
                x0=int(np.round(x[0]+math.floor(m/2)))
                x1=int(np.round(x[1]+math.floor(n/2)))
                if x0<m and x1<n and x0>=0 and x1>=0:
                    output[x0,x1]=img[i,j]
    if retain==False:
        max0=[]
        max1=[]
        for i in range(m):
            for j in range(n):
                x=np.matmul(A,np.array([[i-math.floor(m/2)],[j-math.floor(n/2)]]))
                max0.append(int(np.round(x[0]+math.floor(m/2))))
                max1.append(int(np.round(x[1]+math.floor(n/2))))
        min0=abs(int(min(max0)))
        min1=abs(int(min(max1)))
        print(max(max0)+min0,max(max1)+min1)
        output=np.zeros([int(max(max0)+min0)+1,int(max(max1)+min1)+1])
        for i in range(m):
            for j in range(n):
                x=np.matmul(A,np.array([[i-math.floor(m/2)],[j-math.floor(n/2)]]))
                x0=int(np.round(x[0]+math.floor(m/2)))+min0
                x1=int(np.round(x[1]+math.floor(n/2)))+min1
                if x0>=0 and x1>=0:
                    output[x0,x1]=img[i,j]
                
    return output
```
